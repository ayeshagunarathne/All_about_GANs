This is an implementation of the paper titled:Unpaired Image-to-Image Translation using Cycle-Consistent Adversarial Networks (https://arxiv.org/abs/1703.10593 )

summary of paper:

This paper explains how we can translate an image from one domain to another in the absence of paired image sets.

"We present an approach for learning to translate an image from a source domain X to a target domain Y in the absence of paired examples. Our goal is to learn a mapping G : X → Y such that the distribution of images from G(X) is indistinguishable from the distribution Y using an adversarial loss. Because this mapping is highly under-constrained, we couple it with an inverse mapping F : Y → X and introduce a cycle consistency loss to enforce F(G(X)) ≈ X (and vice versa)." ## Architecture of cycleGAN

image.png

**Figure and the respective caption above is extracted from the original paper**

There are 2 generators: G and F

 **G : X → Y and F : Y → X**

There are 2 adversarial descriminators: DX and DY

**DX aims to distinguish between images {x} and translated images {F(y)}; in the same way, DY aims to discriminate between {y} and {G(x)}**

**Objective function =  Adverasial loss + Cycle consistency loss + identity loss(not used all the time)**

Adversaial loss ▶

\begin{aligned}
\mathcal{L}_{\mathrm{GAN}}\left(G, D_Y, X, Y\right) & =\mathbb{E}_{y \sim p \operatorname{mat}(y)}\left[\log D_Y(y)\right] \\
& +\mathbb{E}_{x \sim p \operatorname{dat}(x)}\left[\log \left(1-D_Y(G(x))\right]\right.
\end{aligned}

\begin{aligned}
\mathcal{L}_{\mathrm{GAN}}\left(F, D_X, Y, X\right) & =\mathbb{E}_{x \sim p \operatorname{mat}(x)}\left[\log D_X(x)\right] \\
& +\mathbb{E}_{y \sim p \operatorname{dat}(y)}\left[\log \left(1-D_X(F(y))\right]\right.
\end{aligned}

Cycle Consistency loss ▶

\begin{aligned}
\mathcal{L}_{\text {cyc }}(G, F) & =\mathbb{E}_{x \sim p_{\text {data }}(x)}\left[\|F(G(x))-x\|_1\right] \\
& +\mathbb{E}_{y \sim p_{\text {dat }}(y)}\left[\|G(F(y))-y\|_1\right]
\end{aligned}

Full objective ▶

\begin{aligned}
\mathcal{L}\left(G, F, D_X, D_Y\right) & =\mathcal{L}_{\mathrm{GAN}}\left(G, D_Y, X, Y\right) \\
& +\mathcal{L}_{\mathrm{GAN}}\left(F, D_X, Y, X\right) \\
& +\lambda \mathcal{L}_{\mathrm{cyc}}(G, F)
\end{aligned}

λ controls the relative importance of the two objectives


---


Final aim is to solve below:

$G^*, F^*=\arg \min _{G, F} \max _{D_x, D_Y} \mathcal{L}\left(G, F, D_X, D_Y\right)$.

As per the authors: This model can be viewed as training two adversarial “autoencoders”
